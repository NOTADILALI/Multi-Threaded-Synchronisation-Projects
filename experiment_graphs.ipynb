```python
#EXPERIMENT-1

import matplotlib.pyplot as plt

sizes = [20, 30, 40, 50, 60, 70, 80, 90, 100]

tas_avg_times = [75, 71, 71, 75, 74, 72, 72, 84, 85]
cas_avg_times = [81, 75, 74, 73, 72, 73, 74, 91, 90]
bcas_avg_times = [73, 72, 76, 75, 73, 76, 96, 84, 93]
seq_avg_times = [75, 72, 74, 72, 73, 72, 77, 76, 83]

plt.figure(figsize=(10, 6))

plt.plot(sizes, tas_avg_times, label="TAS", marker='o')
plt.plot(sizes, cas_avg_times, label="CAS", marker='o')
plt.plot(sizes, bcas_avg_times, label="BCAS", marker='o')
plt.plot(sizes, seq_avg_times, label="SEQ", marker='o')

plt.xlabel("Sudoku Size")
plt.ylabel("Average Time (ms)")
plt.title("Average Time vs Sudoku Size for Different Methods")
plt.legend()
plt.grid(True)

plt.show()
```

```python
#EXPERIMENT-2

import matplotlib.pyplot as plt

# Data (TaskInc values and corresponding average time taken for each method)
taskInc_values = [10, 20, 30, 40, 50]
TAS_times = [74, 73, 73, 72, 73]
CAS_times = [74, 77, 76, 79, 74]
BCAS_times = [82, 73, 74, 76, 75]
SEQ_times = [73, 74, 74, 74, 73]

# Plotting the data
plt.figure(figsize=(8, 6))
plt.plot(taskInc_values, TAS_times, label='TAS', marker='o', color='blue')
plt.plot(taskInc_values, CAS_times, label='CAS', marker='o', color='green')
plt.plot(taskInc_values, BCAS_times, label='BCAS', marker='o', color='red')
plt.plot(taskInc_values, SEQ_times, label='SEQ', marker='o', color='purple')

# Labels and title
plt.xlabel('Task Increment (TaskInc)', fontsize=12)
plt.ylabel('Average Time Taken (ms)', fontsize=12)
plt.title('Time vs Task Increment for Sudoku Validation Methods', fontsize=14)

# Show grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#EXPERIMENT-3

import matplotlib.pyplot as plt

# Hardcoded data for each method (method, number of threads, average time in ms)
threads = [1, 2, 4, 8, 16]  # Only till 16 threads

# Average time taken for each method (excluding 32 threads)
TAS_times = [77, 77, 74, 74, 72]
CAS_times = [83, 81, 74, 74, 75]
BCAS_times = [92, 81, 75, 76, 74]
SEQ_times = [81, 75, 73, 73, 72]

# Create a plot
plt.figure(figsize=(10, 6))

# Plot for each method
plt.plot(threads, TAS_times, label='TAS', marker='o')
plt.plot(threads, CAS_times, label='CAS', marker='o')
plt.plot(threads, BCAS_times, label='BCAS', marker='o')
plt.plot(threads, SEQ_times, label='SEQ', marker='o')

# Set labels and title
plt.xlabel('Number of Threads')
plt.ylabel('Average Time Taken (ms)')
plt.title('Time vs. Number of Threads for Different Methods')

# Display legend
plt.legend(title="Methods")

# Show grid and plot
plt.grid(True)
plt.show()
```
